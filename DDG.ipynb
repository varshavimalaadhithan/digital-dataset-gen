```python
import pandas as pd
import re
from collections import Counter

# Function to process text
def process_text(text):
    text = text.lower()
    words = re.findall(r'\b[a-z]+\b', text)
    word_count = Counter(words)
    return word_count

# Sample input
sample_text = """
AI helps sustainability.
AI helps optimize energy.
Energy efficiency reduces waste.
"""

# Generate dataset
word_data = process_text(sample_text)
df = pd.DataFrame(word_data.items(), columns=["Word", "Frequency"])
df = df.sort_values(by="Frequency", ascending=False)
df.to_csv("dataset_output.csv", index=False)

print("Dataset generated: dataset_output.csv")
df
```
